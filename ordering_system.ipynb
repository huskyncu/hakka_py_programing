```python
from IPython.core.completer import keyword
from prompt_toolkit.filters.cli import vi_digraph_mode
from posixpath import join
from genericpath import samefile
clothes = {"1":"寬衣","2":"寬褲","3":"項鍊","4":"襯衫"}
clothes_sum = {"1":0,"2":0,"3":0,"4":0}
clothes_price = {"1":250,"2":350,"3":300,"4":400 }
choice = ["寬衣250","寬褲350","項鍊300","襯衫400","結束購物"]
total_order = []
total_same = []
sum_l = []
print("4購物車啟動！")
sum=0
while True:
  print("請選擇商品：")
  # 用迴圈印出 (1) 寬衣250  (2) 寬褲350  (3) 項鍊300  (4) 襯衫400  (5) 結束購物  請輸入您的選擇：
  for i in range(len(choice)):
    print("("+str(i+1)+") "+choice[i]," ",end='')
  choice_clothes = input(("請輸入您的選擇："))

  # 如果指令不對就回到一開始重新輸入，continue有回到迴圈一開始地方的功能。
  if choice_clothes !="1" and choice_clothes !="2" and choice_clothes !="3" and choice_clothes !="4" and choice_clothes !="5":
    print("指令不對，請重新輸入")
    continue
  # 如果不是結束購物就加總價錢跟衣服數量
  if choice_clothes!="5":
    sum += clothes_price[choice_clothes]
    clothes_sum[choice_clothes]+=1
  # 以下是印出目前訂了甚麼：
  """
  ex.
  -----------------------------
  目前此客戶買的有：
  -> 寬衣 共 1 件 共 250 元
  -----------------------------
  目前需付： 250 元
  -----------------------------
  """
  print("-----------------------------")
  print("目前此客戶買的有：")
  for k,v in clothes_sum.items():
    if v!=0:
      print("->",clothes[k],"共",v,"件","共",clothes_price[k]*v,"元")
  print("-----------------------------")
  print("目前需付：",sum,"元")
  print("-----------------------------")

  # 如果要加購就回到一開始的地方，不然就跳出迴圈，而裡面有一個是請使用者再三確認訂單，以免發生最後無法修改的處境。
  if choice_clothes!="5":
    con = input("要繼續加購嗎？ (任意鍵) 是，回到選單頁面 (n) 否，結束購物")
    if con=='n':
      print("即將結束購物...")
    else:
      continue
  if con=='n' or choice_clothes=="5":
    con2 = input(("確定結束此客戶訂單？確定後就不能修改了？(y) 是，訂單送出 (任意鍵) 回到選單頁面"))
    if con2=="y":
      break
# 跳出迴圈後，最後印出客戶的訂單，程式結束！
print("-----------------------------")
print("最後訂單：")
for k,v in clothes_sum.items():
  if v!=0:
    print("->",clothes[k],"共",v,"件","共",clothes_price[k]*v,"元")
print("-----------------------------")
print("總共需付：",sum,"元")
print("-----------------------------")

print("訂單已送出！")
```
