```python
import time
import os
from IPython import display
from IPython.display import clear_output
#函式回傳男生跟女生的字串。
def gen(i):          
  if i=="1":
    return "男生"
  elif i=="2":
    return "女生"
#函式回傳bmi
def bmi(height, weight):    
  return weight / (height/100)**2
#函式回傳以bmi判定的體態
def bmi_type(bmi):        
  if bmi>=24 and bmi<27:
    return "過重"
  elif bmi>=27 and bmi<30:
    return "輕度肥胖"
  elif bmi>=30 and bmi<35:
    return "中度肥胖"
  elif bmi>=35:
    return "重度肥胖"
  elif bmi>=18.5 and bmi<24:
    return "適中體重"
  elif bmi<18.5:
    return "過輕"
#函式計算bmr
def bmr(height, weight, gender,age):   
  return 1.2*bmi(height,weight)+ 0.23 * age - 5.4 -10.8*(2-gender)
#函式回傳以bmr判定的體態
def bmr_type(b,gender):            
  if gender == 1:
      if b >=2 and b < 6 :
       return "基礎脂肪"
      elif b >=6 and b < 14:
        return "運動員"
      elif b >=14 and b < 18:
        return "健康"
      elif b >=18 and b < 25:
        return "正常"
      elif b >=25:
       return "肥胖"
  elif gender == 2:
    if b >=10 and b < 14 :
      return "基礎脂肪"
    elif b >=14 and b < 21:
      return "運動員"
    elif b >=21 and b < 25:
      return "健康"
    elif b >=25 and b < 32:
      return "正常"
    elif b >=32:
      return "肥胖"

# loading
#讓=的符號慢慢印出，製造程式有在載入的感覺。
print("載入中...")             
for i in range(50):
  print("=",end='')
  time.sleep(0.05)
print()
print("bmr資料庫系統啟動！")
#藉由while迴圈讓使用者可以一直使用這個服務，按1新增，按2查詢，按3退出程式。
while True:               
  service = input("請選擇服務：新增資料請按1，查詢資料請輸入2，離開程式請按3：")
  #利用break跳出while迴圈以退出程式
  if service =="3":
    print("感謝您的使用！")     
    break
  #新增資料
  elif service =="1":
    #輸入使用者的名字、身分證字號、性別、身高、體重、年紀等資料    
    name = input("請輸入使用者名字：")    
    id = input("請輸入使用者身份證字號：")
    gender = int(input("請選擇性別，男生請輸入1，女生請輸入2："))
    height = int(input("請輸入身高(公分)："))
    weight = int(input("請輸入體重(公斤)："))
    age = int(input("請輸入年齡(單位：歲)："))
    #傳入bmi跟bmr跟兩個判斷體態的式子中並回傳文字跟計算值。
    print("bmi:%5.2f" % bmi(height,weight))
    print("bmi判定體位:"+bmi_type(bmi(height,weight)))
    print("以bmi計算，您的理想體重應為%5.2f公斤至%5.2f公斤之間"%(18.5*(int(height)/100)**2,24*(int(height)/100)**2))
    print("體脂率：%5.2f" % bmr(height,weight,gender,age))
    print("體脂判定體型：%s" % bmr_type(bmr(height,weight,gender,age),gender))
    #新增檔案進資料裡。寫檔
    with open('data.txt', mode='a+') as patient_file:
      patient_file.write(name+" "+id+" "+str(gender)+" "+str(height)+" "+str(weight)+" "+str(age)+"\n")
    print(name+"的資料新增成功")
  #查詢資料
  elif service =="2":
    if os.path.exists("data.txt"):
      name = input("請輸入使用者的名字：")
      id = input("請輸入使用者的身份證字號：")
      #讀檔
      with open('data.txt',mode='r') as patient_file:
        #先讓完檔案所有東西後，再去跑每一行的內容。
        lines = patient_file.readlines()
        for line in lines:
          # 因為讀進來是一整個字串，我把它先換成list，方便我後面的操作。
          line_obj = line.split()
          # 先設定布林值，如果一直沒找到就false。
          has = False
          # 如果name跟id都符合該行資料的第一個資料跟第二個資料，那就執行裡面的程式。
          if line_obj[0]==name and line_obj[1]==id:
            # 找到了。
            has = True
            # 一個一個印出資料
            print("嗨！",line_obj[0])
            print("您的身分證字號："+line_obj[1])
            print("您的性別："+gen(line_obj[2]))
            print("您的身高："+line_obj[3]+"公分")
            print("您的體重："+line_obj[4]+"公斤")
            print("您的年齡："+line_obj[5]+"歲")
            # 我先把list的資料str型態轉型成int型態，這樣方便我後續幾行傳入函式的的操作。
            gender = int(line_obj[2])
            height = int(line_obj[3])
            weight = int(line_obj[4])
            age = int(line_obj[5])
            print("bmi:%5.2f" % bmi(height,weight))
            print("bmi判定體位:"+bmi_type(bmi(height,weight)))
            print("以bmi計算，您的理想體重應為%5.2f公斤至%5.2f公斤之間"%(18.5*(int(height)/100)**2,24*(int(height)/100)**2))
            print("體脂率：%5.2f " % bmr(height,weight,gender,age))
            print("體脂判定體型："+bmr_type(bmr(height,weight,gender,age),gender))
            print("已完成查詢！")
            #因為執行完了，我就跳出迴圈。
            break
        if has == False:
          print("查無此人！")
    else:
      print("沒有資料，請先新增資料！")

  # 接下來就會回到前面，讓使用者自行決定下一步。
  else:
    # 如果有人亂輸指令，就會跑到這行，請他再輸入一次。
    print("請重新輸入指令！")
```
